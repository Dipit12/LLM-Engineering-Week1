```python
from dotenv import load_dotenv

load_dotenv()

import os

api_key = os.getenv("GROQ_API_KEY")

if not api_key:
    print("groq api key not found")
    raise ValueError("groq api key not found")
else:
    print("groq api key found")
```

```python
## making call to an frontier model - Groq
from groq import Groq
messages = [
    {
        "role":"user",
        "content":"What do you know about Major mohit sharma?"
    }
]
client = Groq(api_key = api_key)

response = client.chat.completions.create(
    messages = messages,
    model = "llama-3.3-70b-versatile"
)

print(response.choices[0].message.content)
```

```python
## write a function which calls an given url and extracts the text from the html response
from bs4 import BeautifulSoup
import requests

def extract_text_from_url(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup.get_text()

print(extract_text_from_url("https://www.geeksforgeeks.org/dsa/introduction-to-arrays-data-structure-and-algorithm-tutorials//"))

```

```python
## finally write in a function which takes a url and returns the summary of the url by calling an frontier model

def summarize_url(url):
    text = extract_text_from_url(url)
    messages = [
        {
           "role":"user",
            "content":f"Please summarize the following text: {text}"
        }
    ]
    response = client.chat.completions.create(
        messages = messages,
        model = "llama-3.3-70b-versatile"
    )
    return response.choices[0].message.content

print(summarize_url("https://www.geeksforgeeks.org/dsa/introduction-to-arrays-data-structure-and-algorithm-tutorials//"))
    
```

```python
import requests

url = "https://api.groq.com/openai/v1/chat/completions"
headers = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json"
}

data = {
    "model": "llama-3.3-70b-versatile",
    "messages": [
        {"role": "user", "content": "Tell me a fun fact"}
    ]
}

try:
    response = requests.post(url, headers=headers, json=data)
    res = response.json()

    print(res["choices"][0]["message"]["content"])

except Exception as e:
    print("Error calling the API:", e)
    print("Full response:", response.text)
```
