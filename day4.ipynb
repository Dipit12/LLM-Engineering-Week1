```python
# tokenizing using tiktoken - tokenizer for OpenAI
import tiktoken
```

```python
encoding = tiktoken.encoding_for_model("gpt-4.1-mini")
tokens = encoding.encode("Hi my name is Dipit and i like apple pie")
tokens
```

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id}:{token_text}")
```

## Illusion of memory

```python
import requests
import os
api_key = os.getenv("GROQ_API_KEY")
if not api_key:
    print("API key not found in env file")
    raise ValueError("api key not found")
else:
    print("API key found")
```

```python
from groq import Groq

client = Groq(api_key=api_key)
messages = [
    {"role":"system", "content":"You are a helpful AI assistant"},
    {
        "role":"user", "content":"My name is Dipit , nice to meet you"
    }
]

response = client.chat.completions.create(
    messages=messages,
    model = "llama-3.3-70b-versatile"
)
print(response.choices[0].message.content)
```

```python
## thus all the requests are stateless, thus to create an illusion of memory we keep appending the context in the messages array and pass it as the parameter
# create a simple chatbot
messages = [
    {"role":"system", "content":"You are a helpful AI assistant"},
    {
        "role":"user", "content":"My name is Dipit , nice to meet you"
    }
]
x = True
while(x):
    text = input("What's your message")
    messages.append({"role":"user", "content":f"{text}"})
    response = client.chat.completions.create(
    messages=messages,
    model = "llama-3.3-70b-versatile"
    )
    print(response.choices[0].message.content)
    input_1 = input("Do you want to continue : Y or N")
    if(input_1 == 'N'):
        x = False
```
